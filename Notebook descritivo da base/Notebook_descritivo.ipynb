```python
# Importando as bibliotecas iniciais
import numpy as np
import pandas as pd
```

```python
# Carregando a base e visualizar primeiras linhas
df = pd.read_excel ('/content/Dataset BI MASTER descritivo.xlsx')
# Visualizando a base pelo Explorador de Variáveis
df.head()
```

```python
# Comportamento dos produtos das dezenas
from seaborn.widgets import color_palette
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(50,20))
sns.set(font_scale = 2)
comportamento = sns.lineplot(data=df, y="Produto das dezenas" , x="Ordenamento" , color= "red")
plt.xlabel( "Ordenamento" , size = 24 ) 
plt.ylabel( "Produto das dezenas" , size = 24 ) 
plt.title( "Comportamento" , size = 44) 
plt.show(comportamento)
```

```python
# Comportamento de atributos diversos
plt.figure(figsize=(50,10))
sns.set(font_scale = 2)
comportamento2 = sns.lineplot(data=df, y= "PRECIPITACAO TOTAL, DIARIO(mm) Rio de Janeiro" , x="Ordenamento" , color= "blue")
plt.xlabel( "Ordenamento" , size = 22 ) 
plt.ylabel( "PRECIPITACAO TOTAL, DIARIO(mm) Rio de Janeiro" , size = 22 ) 
plt.title( "Precipitação Total Rio de Janeiro" , size = 35) 
plt.show(comportamento2)
```

```python
# As 10 dezenas mais sorteadas em todos os jogos
freq = pd.DataFrame(df['Coluna1'].tolist() + df['Coluna2'].tolist() + df['Coluna3'].tolist() + df['Coluna4'].tolist() + df['Coluna5'].tolist() + df['Coluna6'].tolist(), columns=['numeros'])
freq['numeros'].value_counts().sort_values(ascending=True).head(10).plot(kind='barh', figsize=(10,5), fontsize=10, legend=False, color='magenta')
plt.title( "As 10 dezenas com maior frequência" , size = 22) 
```
